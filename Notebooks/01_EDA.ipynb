```python
### Conclusions
#- Having at least one funding round may be an indicator of startup success. However, reaching the 3rd or 4th funding round does not add a strong additional signal on its own.
#- Startups with a higher number of funding rounds tend to be more successful.
#- The total number of funding rounds (round count) appears to be a significantly stronger predictive feature than the mere presence of venture capital (VC).
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(startups.isna(), cmap="viridis", cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
### Missing Values
#- Missing values occur in columns like `Unnamed: 6` — a CSV artifact — as well as in `closed_at`, `age_first_milestone_year`, or `age_last_milestone_year`. In these cases, NaNs may indicate that startups did not reach these stages, which itself conveys information.
#- Therefore, it might be better **not to drop these columns**, but instead replace NaN values with 0 where appropriate.
#- Only `Unnamed: 0` and `Unnamed: 6` are removed.
```
